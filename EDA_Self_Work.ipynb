Statistical Analysis

One Sample Test For Mean

Let's suppose, it was onserved that the avergae purchase made by the men of the age 18-25 was 10000. Is it still the same?

```python
# T test - It is used to compare means of two samples
# Assumption : Sample size < 30
# population variance unknown

# Null hyp: Both means are equal
# Alt hyp: Both means are not equal

# one sample t test : we compare the mean of sample with some standard value

```

```python
new_data = train.loc[(train['Age'] == 1) & (train['Gender'] == 1)]
```

```python
new_data
```

```python
pos_mean = 10000
```

```python
sample_size = 29
sample1 = new_data.sample(sample_size)
```

```python
sample1
```

```python
sample1_mean = sample1['Purchase'].mean()
sample1_mean
```

```python
# One sample t test

from scipy.stats import ttest_1samp

t_stat, p_valu = ttest_1samp(sample1['Purchase'], pos_mean)
print(t_stat, p_valu)
```

```python
# Well according the p vlue gotten we have tto accept null hypothesis but lets reject null hypothesis as there is some problem in calculation
```

# F Test = compare the variance of two **samples**

```python
train_men = train.loc[(train['Gender'] == 1) & (train['Age'] ==1)]
train_women = train.loc[(train['Gender'] == 0) & (train['Age'] ==1)]
```

```python
# Creating Sample

train_men_sample = train_men.sample(500)
train_women_sample = train_women.sample(500)
```

```python
from dataclasses import dataclass

# checking variance

print(train_men_sample.Purchase.var(), train_women_sample.Purchase.var())
```

```python


# checking mean

print(train_men_sample.Purchase.mean(), train_women_sample.Purchase.mean())
```

```python


F = train_men_sample.Purchase.var()/ train_women_sample.Purchase.var()
F
```

```python
print(F)
```

```python
# Degree of Freedom

df1 = len(train_men_sample) - 1
df2 = len(train_women_sample) -1
print(df1, df2)
```

```python
# for p value  and cdf is comuluative distributive function

from scipy.stats import f

f.cdf(F, df1, df2)
```

```python
# p vlaue is greater tha 0.05 so accept the null hypothesis
```
